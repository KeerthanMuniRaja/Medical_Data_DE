```python
%sql
CREATE CATALOG IF NOT EXISTS medical_data;
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS medical_data.raw;
```

```python
src_catalog = "workspace"
bronze_catalog = "medical_data"
source_schema = "azure_blob_storage"
raw_schema = "raw"
 
 
 
tables = ["data_dictionary", "patients", "encounters", "organizations", "payers", "procedures"]
 
for table in tables:
    df = spark.table(f"{src_catalog}.{source_schema}.{table}")
   
    df.write.format("delta") \
        .mode("overwrite") \
        .saveAsTable(f"{bronze_catalog}.{raw_schema}.{table}")
 
print("raw Delta tables created")
```
