```python
df_enc = spark.table("medical_data.silver.encounters_s")

dim_encounter = df_enc \
    .withColumnRenamed("id", "encounter_id") \
    .select(
        "encounter_id",
        "encounter_class",
        "code",
        "description",
        "organization"
    ) \
    .dropDuplicates()

dim_encounter.write.mode("overwrite").saveAsTable("medical_data.gold.dim_encounter")
```
